```python
import numpy as np
import pandas as pd
```

```python
edgap = pd.read_csv("edgap_sat_chi.csv")
```

```python
edgap
```

```python
edgap_cleaned = edgap.drop(['LEA_NAME', 'SCH_NAME', 'LCITY', 'LSTATE', 'LAT', 'LON', 'LSTREET1', 'Most Recent Year'],\
                           'columns')
```

```python
edgap_cleaned
```

```python
elsi = pd.read_csv("ELSI_csv.csv")
```

```python
elsi
```

```python
elsi_cleaned = elsi.drop(['School Name', 'State Name [Public School] Latest available year',\
                         'Agency ID - NCES Assigned [Public School] Latest available year',\
                         'ANSI/FIPS State Code [Public School] Latest available year', \
                          'Location ZIP [Public School] 2018-19'],\
                         'columns')
```

```python
elsi_cleaned
```

```python
merged = pd.merge(edgap_cleaned, elsi_cleaned, \
                  right_on='School ID - NCES Assigned [Public School] Latest available year',\
                  left_on='NCESSCH')
```

```python
merged
```

```python
merged.replace({'†' : np.nan, '–' : np.nan, '="0.00"' : np.nan}, inplace=True)
frl = 100 * merged['Free and Reduced Lunch Students [Public School] 2018-19'].astype(float)\
          / merged['Total Students All Grades (Excludes AE) [Public School] 2018-19'].astype(float)
merged['Free and Reduced Lunch Students [Public School] 2018-19'] = 1*frl
```

```python
merged.drop(['School ID - NCES Assigned [Public School] Latest available year', \
             'School Type [Public School] 2018-19', 'Total Students All Grades (Excludes AE) [Public School] 2018-19',\
             'Shared Time School [Public School] 2018-19', 'Type'],\
            'columns', \
            inplace=True)
```

```python
merged.rename(columns={'Charter School [Public School] 2018-19': "Charter School", \
                       'Magnet School [Public School] 2018-19' : "Magnet School", \
                       'Free and Reduced Lunch Students [Public School] 2018-19' : '% Free and Reduced Lunch', \
                       'Grades 9-12 Students [Public School] 2018-19' : '# HS Students',\
                       'Male Students [Public School] 2018-19' : '% Male', \
                       'Female Students [Public School] 2018-19' : '% Female',
                       'Full-Time Equivalent (FTE) Teachers [Public School] 2018-19' : '# FT Teachers',\
                       'Pupil/Teacher Ratio [Public School] 2018-19' : 'Teacher-Student Ratio',\
                       'Score Average' : 'SAT Score Average'}, \
              inplace=True)
merged.astype({'% Free and Reduced Lunch' : float, \
               '# HS Students'            : float, \
               '% Male'                   : float, \
               '% Female'                 : float, \
               '# FT Teachers'            : float, \
               'Teacher-Student Ratio'    : float})
```

```python
m, f = merged['% Male'].astype(float), merged['% Female'].astype(float)
```

```python
merged['% Male']  = m*100 / (m + f)
merged['% Female'] = f*100 / (m + f)
```

```python
merged.replace({'1-Yes' : 1, '2-No' : 0}, inplace=True)
```

```python
merged.to_csv("./EDGAP_ELSI_CombinedTables.csv", index=False)
```
